## Importações das bibliotecas e carregamento dos dados

```python
import pandas as pd
import numpy as np

link_1 = "https://github.com/gabriel7268/Santander_Coders2023.2-Data_Science/raw/main/Datasets/olist_customers_dataset.csv"
link_2 = "https://github.com/gabriel7268/Santander_Coders2023.2-Data_Science/raw/main/Datasets/olist_order_items_dataset.csv"
link_3 = "https://github.com/gabriel7268/Santander_Coders2023.2-Data_Science/raw/main/Datasets/olist_order_payments_dataset.csv"

links = np.array([link_1,link_2,link_3])

df_customers = pd.read_csv(links[0])
df_items = pd.read_csv(links[1])
df_payments = pd.read_csv(links[2])
```

## Tabela de Clientes


```python
display(df_customers.head())
print(f"Shape: {df_customers.shape}")
print(f"Tipo: {type(df_customers)}")
print(f"Tamanho em bytes: {df_customers.to_numpy().nbytes}")
```

Esta tabela, do tipo **Pandas DataFrame**, contém **5 colunas**, as quais dão informações sobre o **ID do cliente, o ID único do cliente, o prefixo do código postal do cliente, a cidade e o estado do cliente**.  

A partir dela podemos identificar que há **99.441 clientes**.


Podemos alterar o nome das colunas, para facilitar a visualização:

```python
print(df_customers.columns)
```

```python
df_customers.columns = ['ID Cliente', 'ID Único Cliente', 'Prefixo Código Postal',
                        'Cidade', 'Estado']
display(df_customers.head())
```

```python
display(df_customers.isnull().sum())
```

A partir da célula acima é possível perceber que **não há dados nulos** em nenhuma linha de nenhuma coluna do DataFrame de clientes.

Vamos, agora, analisar a diferença entre as colunas "ID Cliente" e "ID Único Cliente".

```python
IDUnicoCliente_qnt = df_customers['ID Único Cliente'].unique().size
IDCliente_qnt = df_customers['ID Cliente'].unique().size
print(f"O número de itens únicos da tabela 'ID Cliente' é: {IDCliente_qnt}")
print(f"O número de itens únicos da tabela 'ID Único Cliente' é: {IDUnicoCliente_qnt}")
```

Pode-se perceber que a coluna "ID Cliente" tem 99.441 valores únicos, exatamente igual ao número de clientes, portanto esse ID é realmente único, enquanto que a coluna "ID Único Cliente" tem 96096, pela análise anterior verificamos que não há dados nulos, portanto há, nesta coluna, valores duplicados, portanto eliminares essa coluna da tabela, e trabalharemos apenas com a coluna "ID Cliente" e ela será o indíce da tabela.

```python
df_customers.drop('ID Único Cliente', axis = 1, inplace = True)
df_customers.set_index('ID Cliente', inplace=True)
```

```python
display(df_customers.head())
```

Podemos buscar um cliente específico e saber de qual cidade ele é:

```python
display(df_customers.at['4f2d8ab171c80ec8364f7c12e35b23ad','Cidade'])
```

Podemos ordenar a tabela em ordem alfabética das cidades:

```python
display(df_customers.sort_values('Cidade'))
```

Podemos verificar a quantidade de clientes por cidade, em ordem decrescente para verificar as cidades com maior número de clientes:

```python
display(df_customers.drop('Prefixo Código Postal', axis = 1, inplace = False).groupby('Cidade').value_counts().sort_values(ascending=False))
```

Podemos verificar a quantidade de clientes por estado em ordem decrescente:

```python
display(df_customers.drop(['Prefixo Código Postal','Cidade'], axis = 1, inplace = False).groupby('Estado').value_counts().sort_values(ascending=False))
```

Podemos ver, quais são os estados que tem mais de 1.000 clientes, e quais não.

```python
display(df_customers.drop(['Prefixo Código Postal','Cidade'], axis = 1, inplace = False).groupby('Estado').value_counts() > 1000)
```

Podemos, também, verificar quais são as cidades, do estado RJ, em que há clientes:

```python
display(df_customers[df_customers['Estado'] == 'RJ']['Cidade'].unique())
```

```python
display(df_customers[df_customers['Estado'] == 'RJ']['Cidade'].unique().size)
```

É possível perceber que há, no Rio de Janeiro, 149 cidades em que há clientes.

## Tabela Itens

```python
display(df_items)
print(f"Shape: {df_items.shape}")
print(f"Tipo: {type(df_items)}")
print(f"Tamanho em bytes: {df_items.to_numpy().nbytes}")
```

Esta tabela, por sua vez, contém informações de ID das ordens, Quantidade de itens das ordens, ID dos produtos, ID dos vendedores, data limite de embarque, preço do produto, preço do frete.

```python
print(df_items.columns)
```

Renomeando os nomes das colunas, temos:

```python
df_items.columns = ['ID Ordem', 'Qtd Itens', 'ID Produto', 'ID Vendedor',
       'Data Limite de Embarque', 'Preço', 'Frete']
display(df_items.head())
```

Podemos verificar a quantidade de produtos vendidos por ordem, apresentando a tabela ordenada de forma descendente pela quantidade de itens por ordem.

```python
display(df_items.sort_values('Qtd Itens', ascending = False))
```

Tornando a coluna ID Ordem como índice do DataFrame, temos:

```python
df_items.set_index('ID Ordem', inplace=True)
display(df_items)
```

Podemos analisar a quantidade de itens e o total do valor vendido por cada vendedor.

```python
display(df_items.groupby('ID Vendedor').sum().sort_values('Preço', ascending = False))
```

```python
print(f"A quantidade máxima de produtos vendidos por ordem é: {df_items['Qtd Itens'].max()} produtos.")
print(f"A quantidade mínima de produtos vendidos por ordem é: {df_items['Qtd Itens'].min()} produto.")
print(f"A quantidade média de produtos vendidos por ordem é: {round(df_items['Qtd Itens'].mean(),2)} produto.")
```

```python
print(f"O preço máximo dos produtos vendidos por ordem é: R$ {df_items['Preço'].max()}.")
print(f"O preço mínimo dos produtos vendidos por ordem é: R$ {df_items['Preço'].min()}.")
print(f"O preço médio dos produtos vendidos por ordem é: R$ {round(df_items['Preço'].mean(),2)}.")
```

```python
print(f"O preço máximo do frete por ordem é: R$ {df_items['Frete'].max()}.")
print(f"O preço mínimo do frete por ordem é: R$ {df_items['Frete'].min()}.")
print(f"O preço médio do frete por ordem é: R$ {round(df_items['Frete'].mean(),2)}.")
```

Portanto, é possível notar que há ordens que tem frete grátis. Visto que o valor mínimo do frete é R$ 0.00

## Tabela Pagamentos

```python
display(df_payments)
print(f"Shape: {df_payments.shape}")
print(f"Tipo: {type(df_payments)}")
print(f"Tamanho em bytes: {df_payments.to_numpy().nbytes}")
```

Este DataFrame, possui informações sobre o ID das ordens, o tipo de pagamento, valor do pagamento.

```python
display(df_payments['payment_type'].unique())
```

Com o resultado da saída acima, é possível perceber que os pagamentos utilizados pelos clientes são: Cartão de crédito, boleto, voucher, cartão de débito.

```python
display(df_payments.drop(["order_id", "payment_sequential", "payment_installments",
                          "payment_value"], axis = 1, inplace = False).groupby('payment_type').value_counts().sort_values(ascending = False))
```

Portanto, o método de pagamento mais utilizado pelos clientes é o cartão de crédito, seguido pelo boleto, depois voucher e por fim cartão de débito.
